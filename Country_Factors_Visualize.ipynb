## ***COUNTRY FACTORS VISUALIZATION***
In this mini project we are gonna visualize and see different factors of a country and how they affect or influence each other in a graph format.

We import pandas which is a python library for the data manipulation and analysis of that data.
Matplotlib is imported that is a python library that helps us visualize the data in our graph format.

```
import pandas as pd
import matplotlib.pyplot as plt
```

We import the data from our drive and read the CSV file, convert the CSV file into a dataframe and check the data in that file using df.head()
Data is downloaded from kaggle and the happiness report from 2019 is used.
*URL for the dataset:* "https://www.kaggle.com/unsdsn/world-happiness"





```
country_data = pd.read_csv("/content/drive/My Drive/Dataset/2019.csv")
country_data.head()
```

First we are gonna visualize the happiness score of each country using a bar graph and see how the happiness score has varied over different countries.
Here we can see **Finland** has the hightest happiness score while **South Sadan** has the least happiness score 

```
plt.rcParams.update({'font.size':'30'})
country_happiness_plot = country_data.plot(kind="bar",x="Country",y="Score",color="blue",figsize=(65,20))
country_happiness_plot.set_xlabel("NAME OF COUNTRY")
country_happiness_plot.set_ylabel("SCORE")
plt.tick_params(labelsize=23)
plt.show()
```

Now we will visualize how different factors are related to each other. Using graph we will try to visualize the that relation that we can set with different factors using line graph and try to see some relations between them.

When we visualize relation between *Happiness Score* and *GDP per capita* we can see as the GDP is high so is the happiness score of the country. There are variation in the middle but if we see in the start and end we can see more the GDP more the happiness.

```
plt.rcParams.update({'font.size':'15'})
gdp_happiness_plot = country_data.plot(kind="line",x="Score",y="GDP per capita", color="orange", figsize=(35,10))
gdp_happiness_plot.set_xlabel("SCORE")
gdp_happiness_plot.set_ylabel("GDP PER CAPITA")
plt.show()
```

This graph between Happiness Score and Perception of corruption shows that the perception is not much related because the variation in every level of score is a lot.

```
plt.rcParams.update({'font.size':'15'})
happiness_corruption_plot = country_data.plot(kind="line",x="Score",y="Perception of Corruption", color="green", figsize=(35,10))
happiness_corruption_plot.set_xlabel("SCORE")
happiness_corruption_plot.set_ylabel("PERCEPTION OF CORRUPTION")
plt.show()
```

This graph can clearly show with the increasing happiness score the freedom of choices is increasing or vice versa. The more the people have their freedom more are they happy. More the freedom more we can incease the happiness score.

```
plt.rcParams.update({'font.size':'15'})
happiness_freedom_plot = country_data.plot(kind="line",x="Score",y="Freedom for choices", color="red", figsize=(35,10))
happiness_freedom_plot.set_xlabel("SCORE")
happiness_freedom_plot.set_ylabel("FREEDOM FOR CHOICES")
plt.show()
```

The increase in life expectency has clearly shown that there is increase in happiness score. More the people are healthy more the people are happy. Although there are few variations.

```
plt.rcParams.update({'font.size':'15'})
happiness_life_plot = country_data.plot(kind="line",x="Score",y="Healthy life expectency", color="green", figsize=(35,10))
happiness_life_plot.set_xlabel("SCORE")
happiness_life_plot.set_ylabel("HEALTHY LIFE EXPECTENCY")
plt.show()
```

This graph between the GDP per capita and Perception of corruption clearly shows that there is minimum to no relation between GDP and corruption or rather it is a matter of choice.

```
plt.rcParams.update({'font.size':'15'})
gdp_corruption_plot = country_data.plot(kind="line",x="GDP per capita",y="Perception of Corruption", color="blue", figsize=(35,10))
gdp_corruption_plot.set_xlabel("GDP PER CAPITA")
gdp_corruption_plot.set_ylabel("PERCEPTION OF CORRUPTION")
plt.show()
```

Even with the GDP per capita and Healthy life expectency using line graph we cannot determine a relation between these two factors. We will need to use other graph to gain more insight on these. 

```
plt.rcParams.update({'font.size':'15'})
gdp_life_plot = country_data.plot(kind="line",x="GDP per capita",y="Healthy life expectency", color="red", figsize=(35,10))
gdp_life_plot.set_xlabel("GDP PER CAPITA")
gdp_life_plot.set_ylabel("HEALTHY LIFE EXPECTENCY")
plt.show()
```
